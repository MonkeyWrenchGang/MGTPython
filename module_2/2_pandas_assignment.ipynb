## Query 0. 

Create a query that returns the eventid, eventtype,borough, startdatetime, enddatetime, country,category, subcategoryname,    where country equals Japan and borough equals Manhattan. 


**sample answer:**
```python
# -- DataFrame Filtering -- 
q0 = df[(df["country"] == "Japan") & (df["borough"] == "Manhattan")] \
[["eventid", "eventtype","borough", "startdatetime", "enddatetime", "country","category", "subcategoryname"]]


# -- OR -- 
q0 = df.query(' country == "Japan" and borough == "Manhattan"')\
    [["eventid", "eventtype","borough", "startdatetime", "enddatetime", "country","category", "subcategoryname"]]

# -- show query result -- 
q0
```
