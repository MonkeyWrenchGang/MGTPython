# Jazzing up your seaborn plot game! 


---

Seaborn offers a wide range of built-in styles and  customization options to jazz up your Seaborn plot game and take your visualizations to the next level. In this notebook we'll cover the top 5 or 6 ways to take your plot game up a level:

1. Supressing information text 
2. Changing the figure size
3. Move legends around
4. Set Plot Style
5. Use Subplots 
  - use facet grid 
6. Color Palettes 

````python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Import data
df = sns.load_dataset('penguins').rename(columns={'sex': 'gender'})
df.head()
````

# 1. suppress this text output 


---

By default, seaborn will output some text after each plot like `<matplotlib.axes._subplots.AxesSubplot at 0x7fd11d58eb50>` the text is just junk and makes your plot look less-professional. Let's see how we can supress this. 

> NOTE: For most of these examples we are going to use a standard scatterplot and the built-in `penguins` dataset 



````python
# standard Scatter Plot
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender')
plt.title("this is a title");
````

Did you notice the text output just above the plot? something like this... 

```
<matplotlib.axes._subplots.AxesSubplot at 0x7f3ba9735340>
```

To suppress this text output simply add a  `;` at the end of your plot. like this

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender');
```

````python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender');
````

If you add a title or something after the plot simply place a `;` after the last statement. Otherwise you'll get something like `Text(0.5, 1.0, 'Standard Scatterplot')` as output. The `;` should go at the end of the last statement. 

````python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender');
plt.title("Standard Scatterplot") # ;
````

#3. Move Legends around 

There are two components to move legends around loc and bbox_to_anchor(). 

- 1st loc = location determines general location like :
  -'best': automatically choose the best position for the legend
  - 'upper right': place the legend in the upper-right corner of the plot
  - 'upper left': place the legend in the upper-left corner of the plot
  - 'lower right': place the legend in the lower-right corner of the plot
  - 'lower left': place the legend in the lower-left corner of the plot
  - 'lower center': place the legend in the center of the bottom of the plot
- 'upper center': place the legend in the center of the top of the plot

- 2nd bbox_to_anchor=() is used to adjust the position of a legend with respect to a plot. 
- x: the horizontal position of the bounding box
- y: the vertical position of the bounding box

```python
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1));
```

````python
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species')
plt.show()

#place legend outside top right corner of plot
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
plt.title("top right corner of plot");
plt.show()
#place legend outside center right border of plot
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species')
plt.legend(bbox_to_anchor=(1.02, 0.55), loc='upper left', borderaxespad=0);
plt.title("center right border of plot");
plt.show()
#place legend outside bottom right corner of plot
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species')
plt.legend(bbox_to_anchor=(1.02, 0.15), loc='upper left', borderaxespad=0);
plt.title("bottom right corner of plot");
````

````python
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender')
plt.legend(loc='best', 
           #bbox_to_anchor=(0,1)
           );

````

#4. Set the plot style

## using: sns.set_style()


---


The set_style() function is used to set the aesthetic style of the plots. The set_style() function can be used to set the default style or context for all the plots in the current script or notebook. It takes one required argument, style, which specifies the name of the style to use.

Here are options to try out:
- 'darkgrid': a dark gray background with grid lines
- 'whitegrid': a white background with grid lines
- 'dark': a dark gray background without grid lines
- 'white': a white background without grid lines
- 'ticks': a white background with tick marks along the axes

You can also use the set_context() function to set the context of the plot, which determines the size and scale of the plot elements. Here are some available contexts:

- 'paper': smaller font sizes and thinner lines, suitable for publications
- 'notebook': the default context with slightly larger font sizes
- 'talk': larger font sizes and thicker lines, suitable for presentations
- 'poster': very large font sizes and thick lines, suitable for posters

````python
# Change default style
sns.set_style('whitegrid')
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.title("set_style('whitegrid')");
````

````python
# Change default style
sns.set_style('darkgrid')
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.title("set_style('darkgrid')");
````

````python
# Change default style
sns.set_style('ticks')
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.title("set_style('ticks')");
````

# 5. Subplots


---

One useful feature of Matplotlib is its ability to create multiple plots, called subplots, within a single figure.

The subplots() function can be used to create multiple plots in the same figure. The subplots() function returns a figure object and a set of axes objects, which can be used to create and customize the individual plots.


To create subplots in Matplotlib, you can use the `plt.subplots()` function. This function returns a figure object and an array of `axes objects`, which can be used to create and manipulate subplots.

`plt.subplots(nrows=2, ncols=2, figsize=(8, 4))`

- nrows: number of rows
- ncols: number of columns 
- figsize: total figure size 

Here's an example code snippet to create a 2x2 grid of subplots:

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))


```

````python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(5, 5))
````

## Let's take it a step further.


---

Here let's see how we can the subplots to our figure. to do this we simply tell seaborn what axis (`ax`) we want to place the plot in. 



````python

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5, 10))

sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', 
                ax=axes[0])

sns.boxplot(data=df, x='body_mass_g', y='species', 
            ax=axes[1]);
````

````python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5));

sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                alpha=0.7, hue='species', ax=axes[0])
sns.boxplot(data=df, x='body_mass_g', y='species', ax=axes[1]);
````

## 5a. Facet Grid

You can also use the sns.FacetGrid() function. This function returns a FacetGrid object, which can be used to create and manipulate subplots.

Here's an example code snippet to create a 2x2 grid of subplots:


## Custom Color Palettes


---

In addition to the built-in color palettes, Seaborn also allows you to create custom color palettes to use in your visualizations.

To create a custom color palette in Seaborn, you can simply pass a dictionary of colors mapped to a categorical variable. Additionally you can use the `sns.color_palette()` function to change the default color palette by passing a list of colors to the function.

For example, to create a custom palette with three colors (red, green, and blue) in hexidecimal format like this:

```python
import seaborn as sns

my_palette = sns.color_palette(["#FF0000", "#00FF00", "#0000FF"])

```

OR simply pass a dictionary mapping where a categorical is mapped to a color like this: 

```
my_palette = {"Dinner": "red",
              "Lunch": "lightblue"}

sns.countplot(data=tips, 
            x="time",
            palette=my_palette
            )
plt.title("count of times w. custom color palette");
plt.show();


my_palette = {"Dinner": "red",
              "Lunch": "lightblue"}

sns.countplot(data=tips, 
            x="time",
            palette=my_palette
            )
plt.title("count of times w. custom color palette");
plt.show();

```

````python
# Import data
tips = sns.load_dataset('tips')
tips["day"].value_counts()
````

````python


# -----
my_palette = {"Dinner": "orange",
              "Lunch": "green"}

sns.countplot(data=tips, 
            x="time",
            palette=my_palette
            )
plt.title("count of times w. custom color palette");
plt.show();

# -----
my_palette = {"Dinner": "red",
              "Lunch": "lightblue"}

sns.countplot(data=tips, 
            x="time",
            palette=my_palette
            )
plt.title("count of times w. custom color palette");
plt.show();

# -----
my_palette = {"Dinner": "#0000FF",
              "Lunch": "#00FF00"}

sns.countplot(data=tips, 
            x="time",
            palette=my_palette
            )
plt.title("count of times w. custom color palette");
plt.show();
````
