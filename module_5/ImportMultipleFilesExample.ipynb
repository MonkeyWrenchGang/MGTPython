# How do i import multiple files into a panda's dataframe?


---



```python
# --------------------------------
import warnings
warnings.filterwarnings('ignore')
# --------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns

%matplotlib inline
```

# Looping to load a bunch of files from Github 


---



1. start with the base path to the files
2. create an empty data frame, this is where we are going to append our imported data to. 
3. loop over the files, in this case we are using a range function to give us a sequence number range(1,4) will produce 1, 2, and 3 but not 4. 
4. we'll create a string that contains the full path to the file using the "format()" function. 
5. next we'll import the file to a "temporary" dataframe
6. we'll finally append the file to our empty data frame
7. repeat for each file. 



```python
# base file path. 
path = "https://raw.githubusercontent.com/MonkeyWrenchGang/MGTPython/main/module_5/mid-term/"

# create empty data frame 
phys = pd.DataFrame()

for i in range(1,4):
  #print(i)
  file = "{}scnctn_physicians_{}.csv".format(path,i)
  print(file)
  # import the file 
  temp = pd.read_csv(file)
  # here you need to concatenate 
  phys = pd.concat([phys,temp])

print("file contains {} rows and {} columns".format(phys.shape[0],phys.shape[1]))
```

# Clean up Column Names and Check Rows and Columns 


---



```python
phys.columns = ( phys.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
    .str.replace('(', '')
    .str.replace(')', '')
    .str.replace('?', '')
    .str.replace('\'', '') # notice the backslash \ this is an escape character
)
print(phys.columns)
phys.head()
```
